La standardisation des données est une technique puissante utilisée en science des données pour rendre les variables comparables entre elles, tout en réduisant l'impact des valeurs aberrantes et   aide également à atténuer l'impact des valeurs aberrantes, ces points de données extrêmes qui peuvent fausser les résultats de vos analyses. Comment cela fonctionne-t-il ? Eh bien, la standardisation calcule la moyenne et l'écart-type des variables, et utilise ces informations pour centrer les données autour de zéro et les mettre à l'échelle de manière appropriée. En conséquence, les valeurs aberrantes, qui sont souvent très éloignées de la moyenne, sont réduites à des valeurs plus raisonnables.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Création d'un objet MinMaxScaler
scaler = MinMaxScaler()

# Adapter et transformer les données d'entraînement
X_train_scaled = scaler.fit_transform(X_train)

# Transformer les données de test en utilisant les paramètres de l'entraînement
X_test_scaled = scaler.transform(X_test)
```



***En premier lieu, il est crucial de mentionner les métriques d'évaluation les plus importantes lors de ce projet, car mon jeu de données est très crucial. Dans le contexte spécifique du diabète, où la détection précise des cas positifs (personnes diabétiques) est essentielle pour un traitement précoce et approprié, le rappel (recall) pourrait être une métrique d'évaluation particulièrement importante. Cependant, il est toujours recommandé d'examiner plusieurs métriques pour obtenir une vue d'ensemble de la performance du modèle, telles que l'accuracy et la matrice de confusion***





pour l'application de logistic regression dans le cas de multiclass classification il  utilise  des stratégies one vs all est one vs one pour bien pour les transformer en classification binnaire dans lequele pour le choix de classe pour une nouvelle données il faut faire un vote majorutaire et respecter quelques caractéristique tels que :

One-vs-Rest (One-vs-All) :

Avantages :

    *  Simple à implémenter et à comprendre.
    *  Convient bien aux ensembles de données avec un grand nombre de classes.
       Performant lorsque les classes sont bien séparables et qu'il n'y a pas de chevauchement significatif entre les classes.
    * Facile à interpréter les résultats, car chaque classe est évaluée par rapport au reste des classes.

Inconvénients :

    Peut être inefficace lorsque les classes sont déséquilibrées ou que les chevauchements entre les classes sont importants.
    Les prédictions peuvent être biaisées vers les classes majoritaires, en particulier si les classes sont déséquilibrées.


One-vs-One :

Avantages :

    *   Peut être plus efficace lorsque les classes sont déséquilibrées ou qu'il y a des chevauchements significatifs entre les classes.
    *   Peut offrir de meilleures performances pour les ensembles de données avec un petit nombre de classes.
    *   Peut mieux gérer les cas où les classes sont difficiles à séparer et  oins sensible aux déséquilibres de classe

Inconvénients :

    Plus coûteux en termes de calculs et de mémoire, car il nécessite l'entraînement de plusieurs classifieurs binaires.
    Plus complexe à interpréter les résultats, car les prédictions sont basées sur des comparaisons entre paires de classes.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import  classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score

# Liste des stratégies à évaluer
strategies = ["ovr", "multinomial"]

for strategy in strategies:
    # Création du modèle en utilisant la stratégie actuelle
    model = LogisticRegression(multi_class=strategy)   #L'option average='macro' calcule le rappel pour chaque classe individuellement, puis renvoie la moyenne non pondérée de ces rappels. Cette option donne le même poids à chaque classe, quelle que soit sa taille.

    # Entraînement du modèle
    model.fit(X_train_scaled, y_train)

    # Faire des prédictions
    y_pred = model.predict(X_test_scaled)

    # Évaluation du modèle
    recall= recall_score(y_test, y_pred,average='macro')   #L'option average='macro' calcule le rappel pour chaque classe individuellement, puis renvoie la moyenne non pondérée de ces rappels. Cette option donne le même poids à chaque classe, quelle que soit sa taille.

    print(f"Le rappel du modèle ({strategy}):", recall)

    # Rapport de classification
    print(f"Rapport de classification ({strategy}):")
    print(classification_report(y_test, y_pred))

    # Calcul de la matrice de confusion
    conf_matrix = confusion_matrix(y_test, y_pred)

    # Affichage de la matrice de confusion sous forme de heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",  xticklabels=model.classes_, yticklabels=model.classes_ )
    plt.xlabel('Valeurs prédites')
    plt.ylabel('Valeurs réelles')
    plt.title('Matrice de confusion')
    plt.show()
    print("-" * 100)
```

 L'objectif principal de la phase de validation est de s'assurer que le modèle généralise bien aux données qu'il n'a pas encore vues. Cela permet d'évaluer l'aptitude du modèle à faire des prédictions précises sur de nouvelles données non utilisées pendant l'entraînement.

```python
# Diviser les données d'entraînement en ensembles d'entraînement et de validation (50% chacun)
X_train1, X_val, y_train1, y_val = train_test_split(X_train_scaled, y_train, test_size=0.4, random_state=42)

for strategy in strategies:

    model = LogisticRegression(multi_class=strategy)

    # Entraînement du modèle sur l'ensemble d'entraînement
    model.fit(X_train1, y_train1)

    # Faire des prédictions sur l'ensemble de validation
    y_val_pred = model.predict(X_val)

    # Évaluation du modèle sur l'ensemble de validation
    recall= recall_score(y_val, y_val_pred,average='macro')

    print(f"Rappel sur l'ensemble de validation ({strategy}):",recall)

    # Si le modèle semble performant sur l'ensemble de validation, vous pouvez ensuite évaluer ses performances sur l'ensemble de test

    seuil_de_performance=0.85
    if recall > seuil_de_performance:
        # Faire des prédictions sur l'ensemble de test
        y_test_pred = model.predict(X_test_scaled)

        # Évaluation du modèle sur l'ensemble de test
        recall= recall_score(y_test, y_test_pred,average='macro')
        print(f"Rappel sur l'ensemble de test ({strategy}):", recall)


        # Matrice de confusion sur l'ensemble de test
        conf_matrix_test = confusion_matrix(y_test, y_test_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(conf_matrix_test, annot=True, fmt="d", cmap="Blues", xticklabels=model.classes_, yticklabels=model.classes_)
        plt.xlabel('Valeurs prédites')
        plt.ylabel('Valeurs réelles')
        plt.title('Matrice de confusion sur l\'ensemble de test')
        plt.show()
        print("-" * 100)
```

choisir la meilleure valeur de k en utilisant la méthode du coude (elbow method), créer un modèle k-NN, faire des prédictions et évaluer la performance du modèle sur les ensembles d'entraînement et de test.

La valeur optimale de k serait celle où l'exactitude sur les données de test est maximale sans surapprentissage (l'écart entre les scores d'entraînement et de test est faible

```python
from sklearn.neighbors import KNeighborsClassifier


# Liste des valeurs de k à tester
k_values = range(1, 25)

# Liste pour stocker les scores d'entraînement et de test
train_scores = []
test_scores = []

# Boucle sur les différentes valeurs de k
for k in k_values:
    # Initialisation du modèle k-NN avec k voisins
    knn = KNeighborsClassifier(n_neighbors=k)

    # Entraînement du modèle sur les données d'entraînement
    knn.fit(X_train, y_train)

    # Calcul de l'exactitude sur les données d'entraînement
    train_score = knn.score(X_train, y_train)
    train_scores.append(train_score)

    # Calcul de l'exactitude sur les données de test
    test_score = knn.score(X_test, y_test)
    test_scores.append(test_score)

# Affichage des scores d'entraînement et de test
plt.plot(k_values, train_scores, label='Train Score')
plt.plot(k_values, test_scores, label='Test Score')
plt.xlabel('Valeur de k')
plt.ylabel('Exactitude')
plt.title('Méthode du coude pour choisir k')
plt.legend()
plt.show()
```

```python
# Choisissez la meilleure valeur de k
best_k = [12 , 13 ,15 ]
for i in  best_k:
  print("pour k egale a :",i)
  # Création du modèle k-NN avec la meilleure valeur de k
  best_knn_model = KNeighborsClassifier(n_neighbors=i)

  # Entraînement du modèle sur l'ensemble d'entraînement
  best_knn_model.fit(X_train, y_train)

  # Prédictions sur l'ensemble de test
  y_pred = best_knn_model.predict(X_test)

  # Matrice de confusion
  conf_matrix = confusion_matrix(y_test, y_pred)
  print("Matrice de confusion :\n", conf_matrix)

  # Rapport de classification
  print("\nRapport de classification :\n", classification_report(y_test, y_pred))

  # Exactitude
  accuracy = best_knn_model.score(X_test, y_test)
  print("\nExactitude :", accuracy)

  print("=======================================================================")
```

```python
from sklearn.model_selection import cross_val_score

# Liste des valeurs de k à tester
k_values = range(1, 25)

# Initialisation d'une liste pour stocker les scores moyens de validation croisée
cv_scores = []

# Boucle sur les différentes valeurs de k
for k in k_values:
    # Initialisation du modèle k-NN avec k voisins
    knn = KNeighborsClassifier(n_neighbors=k)

    # Calcul des scores de validation croisée
    scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')

    # Calcul de la moyenne des scores de validation croisée
    cv_scores.append(scores.mean())

# Trouver la meilleure valeur de k avec la meilleure précision de validation croisée
best_k_index = cv_scores.index(max(cv_scores))
best_k = k_values[best_k_index]

# Affichage de la meilleure valeur de k
print("Meilleure valeur de k :", best_k)

# Affichage des scores moyens de validation croisée
plt.plot(k_values, cv_scores)
plt.xlabel('Valeur de k')
plt.ylabel('Score de validation croisée')
plt.title('Validation croisée pour sélectionner k')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
# Définition de la grille des hyperparamètres pour le noyau RBF
param_grid_rbf = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 1]
}

# Initialisation du modèle SVM avec le noyau RBF
svm_rbf_model = SVC(kernel='rbf')

# Recherche sur grille pour trouver les meilleurs hyperparamètres pour le noyau RBF
grid_search_rbf = GridSearchCV(estimator=svm_rbf_model, param_grid=param_grid_rbf, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_rbf.fit(X_train, y_train)

# Affichage des meilleurs hyperparamètres trouvés pour le noyau RBF
print("Meilleurs hyperparamètres pour le noyau RBF :", grid_search_rbf.best_params_)

# Choix du meilleur modèle pour le noyau RBF
best_svm_rbf_model = grid_search_rbf.best_estimator_

# Prédiction sur l'ensemble de test avec le meilleur modèle pour le noyau RBF
y_pred_rbf = best_svm_rbf_model.predict(X_test)

# Évaluation de la performance du modèle pour le noyau RBF
accuracy_rbf = best_svm_rbf_model.score(X_test, y_test)
print("Précision pour le noyau RBF :", accuracy_rbf)
```
