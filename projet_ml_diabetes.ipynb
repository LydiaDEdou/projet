La standardisation des donnes est une technique puissante utilise en science des donnes pour rendre les variables comparables entre elles, tout en rduisant l'impact des valeurs aberrantes et   aide galement  attnuer l'impact des valeurs aberrantes, ces points de donnes extrmes qui peuvent fausser les rsultats de vos analyses. Comment cela fonctionne-t-il ? Eh bien, la standardisation calcule la moyenne et l'cart-type des variables, et utilise ces informations pour centrer les donnes autour de zro et les mettre  l'chelle de manire approprie. En consquence, les valeurs aberrantes, qui sont souvent trs loignes de la moyenne, sont rduites  des valeurs plus raisonnables.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Division des donnes en ensembles d'entranement et de test
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Cration d'un objet MinMaxScaler
scaler = MinMaxScaler()

# Adapter et transformer les donnes d'entranement
X_train_scaled = scaler.fit_transform(X_train)

# Transformer les donnes de test en utilisant les paramtres de l'entranement
X_test_scaled = scaler.transform(X_test)
```

pour l'application de logistic regression dans le cas de multiclass classification il  utilise  des stratgies one vs all est one vs one pour bien pour les transformer en classification binnaire dans lequele pour le choix de classe pour une nouvelle donnes il faut faire un vote majorutaire et respecter quelques caractristique tels que :

One-vs-Rest (One-vs-All) :

Avantages :

    *  Simple  implmenter et  comprendre.
    *  Convient bien aux ensembles de donnes avec un grand nombre de classes.
       Performant lorsque les classes sont bien sparables et qu'il n'y a pas de chevauchement significatif entre les classes.
    * Facile  interprter les rsultats, car chaque classe est value par rapport au reste des classes.

Inconvnients :

    Peut tre inefficace lorsque les classes sont dsquilibres ou que les chevauchements entre les classes sont importants.
    Les prdictions peuvent tre biaises vers les classes majoritaires, en particulier si les classes sont dsquilibres.


One-vs-One :

Avantages :

    *   Peut tre plus efficace lorsque les classes sont dsquilibres ou qu'il y a des chevauchements significatifs entre les classes.
    *   Peut offrir de meilleures performances pour les ensembles de donnes avec un petit nombre de classes.
    *   Peut mieux grer les cas o les classes sont difficiles  sparer et  oins sensible aux dsquilibres de classe

Inconvnients :

    Plus coteux en termes de calculs et de mmoire, car il ncessite l'entranement de plusieurs classifieurs binaires.
    Plus complexe  interprter les rsultats, car les prdictions sont bases sur des comparaisons entre paires de classes.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import  classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score

# Liste des stratgies  valuer
strategies = ["ovr", "multinomial"]

for strategy in strategies:
    # Cration du modle en utilisant la stratgie actuelle
    model = LogisticRegression(multi_class=strategy)   #L'option average='macro' calcule le rappel pour chaque classe individuellement, puis renvoie la moyenne non pondre de ces rappels. Cette option donne le mme poids  chaque classe, quelle que soit sa taille.

    # Entranement du modle
    model.fit(X_train_scaled, y_train)

    # Faire des prdictions
    y_pred = model.predict(X_test_scaled)

    # valuation du modle
    recall= recall_score(y_test, y_pred,average='macro')   #L'option average='macro' calcule le rappel pour chaque classe individuellement, puis renvoie la moyenne non pondre de ces rappels. Cette option donne le mme poids  chaque classe, quelle que soit sa taille.

    print(f"Le rappel du modle ({strategy}):", recall)

    # Rapport de classification
    print(f"Rapport de classification ({strategy}):")
    print(classification_report(y_test, y_pred))

    # Calcul de la matrice de confusion
    conf_matrix = confusion_matrix(y_test, y_pred)

    # Affichage de la matrice de confusion sous forme de heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",  xticklabels=model.classes_, yticklabels=model.classes_ )
    plt.xlabel('Valeurs prdites')
    plt.ylabel('Valeurs relles')
    plt.title('Matrice de confusion')
    plt.show()
    print("-" * 100)
```

 L'objectif principal de la phase de validation est de s'assurer que le modle gnralise bien aux donnes qu'il n'a pas encore vues. Cela permet d'valuer l'aptitude du modle  faire des prdictions prcises sur de nouvelles donnes non utilises pendant l'entranement.

```python
# Diviser les donnes d'entranement en ensembles d'entranement et de validation (50% chacun)
X_train1, X_val, y_train1, y_val = train_test_split(X_train_scaled, y_train, test_size=0.4, random_state=42)

for strategy in strategies:

    model = LogisticRegression(multi_class=strategy)

    # Entranement du modle sur l'ensemble d'entranement
    model.fit(X_train1, y_train1)

    # Faire des prdictions sur l'ensemble de validation
    y_val_pred = model.predict(X_val)

    # valuation du modle sur l'ensemble de validation
    recall= recall_score(y_val, y_val_pred,average='macro')

    print(f"Rappel sur l'ensemble de validation ({strategy}):",recall)

    # Si le modle semble performant sur l'ensemble de validation, vous pouvez ensuite valuer ses performances sur l'ensemble de test

    seuil_de_performance=0.85
    if recall > seuil_de_performance:
        # Faire des prdictions sur l'ensemble de test
        y_test_pred = model.predict(X_test_scaled)

        # valuation du modle sur l'ensemble de test
        recall= recall_score(y_test, y_test_pred,average='macro')
        print(f"Rappel sur l'ensemble de test ({strategy}):", recall)


        # Matrice de confusion sur l'ensemble de test
        conf_matrix_test = confusion_matrix(y_test, y_test_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(conf_matrix_test, annot=True, fmt="d", cmap="Blues", xticklabels=model.classes_, yticklabels=model.classes_)
        plt.xlabel('Valeurs prdites')
        plt.ylabel('Valeurs relles')
        plt.title('Matrice de confusion sur l\'ensemble de test')
        plt.show()
        print("-" * 100)
```

choisir la meilleure valeur de k en utilisant la mthode du coude (elbow method), crer un modle k-NN, faire des prdictions et valuer la performance du modle sur les ensembles d'entranement et de test.

La valeur optimale de k serait celle o l'exactitude sur les donnes de test est maximale sans surapprentissage (l'cart entre les scores d'entranement et de test est faible

```python
from sklearn.neighbors import KNeighborsClassifier


# Liste des valeurs de k  tester
k_values = range(1, 25)

# Liste pour stocker les scores d'entranement et de test
train_scores = []
test_scores = []

# Boucle sur les diffrentes valeurs de k
for k in k_values:
    # Initialisation du modle k-NN avec k voisins
    knn = KNeighborsClassifier(n_neighbors=k)

    # Entranement du modle sur les donnes d'entranement
    knn.fit(X_train, y_train)

    # Calcul de l'exactitude sur les donnes d'entranement
    train_score = knn.score(X_train, y_train)
    train_scores.append(train_score)

    # Calcul de l'exactitude sur les donnes de test
    test_score = knn.score(X_test, y_test)
    test_scores.append(test_score)

# Affichage des scores d'entranement et de test
plt.plot(k_values, train_scores, label='Train Score')
plt.plot(k_values, test_scores, label='Test Score')
plt.xlabel('Valeur de k')
plt.ylabel('Exactitude')
plt.title('Mthode du coude pour choisir k')
plt.legend()
plt.show()
```

```python
# Choisissez la meilleure valeur de k
best_k = [12 , 13 ,15 ]
for i in  best_k:
  print("pour k egale a :",i)
  # Cration du modle k-NN avec la meilleure valeur de k
  best_knn_model = KNeighborsClassifier(n_neighbors=i)

  # Entranement du modle sur l'ensemble d'entranement
  best_knn_model.fit(X_train, y_train)

  # Prdictions sur l'ensemble de test
  y_pred = best_knn_model.predict(X_test)

  # Matrice de confusion
  conf_matrix = confusion_matrix(y_test, y_pred)
  print("Matrice de confusion :\n", conf_matrix)

  # Rapport de classification
  print("\nRapport de classification :\n", classification_report(y_test, y_pred))

  # Exactitude
  accuracy = best_knn_model.score(X_test, y_test)
  print("\nExactitude :", accuracy)

  print("=======================================================================")
```

```python
from sklearn.model_selection import cross_val_score

# Liste des valeurs de k  tester
k_values = range(1, 25)

# Initialisation d'une liste pour stocker les scores moyens de validation croise
cv_scores = []

# Boucle sur les diffrentes valeurs de k
for k in k_values:
    # Initialisation du modle k-NN avec k voisins
    knn = KNeighborsClassifier(n_neighbors=k)

    # Calcul des scores de validation croise
    scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')

    # Calcul de la moyenne des scores de validation croise
    cv_scores.append(scores.mean())

# Trouver la meilleure valeur de k avec la meilleure prcision de validation croise
best_k_index = cv_scores.index(max(cv_scores))
best_k = k_values[best_k_index]

# Affichage de la meilleure valeur de k
print("Meilleure valeur de k :", best_k)

# Affichage des scores moyens de validation croise
plt.plot(k_values, cv_scores)
plt.xlabel('Valeur de k')
plt.ylabel('Score de validation croise')
plt.title('Validation croise pour slectionner k')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
# Dfinition de la grille des hyperparamtres pour le noyau RBF
param_grid_rbf = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 1]
}

# Initialisation du modle SVM avec le noyau RBF
svm_rbf_model = SVC(kernel='rbf')

# Recherche sur grille pour trouver les meilleurs hyperparamtres pour le noyau RBF
grid_search_rbf = GridSearchCV(estimator=svm_rbf_model, param_grid=param_grid_rbf, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_rbf.fit(X_train, y_train)

# Affichage des meilleurs hyperparamtres trouvs pour le noyau RBF
print("Meilleurs hyperparamtres pour le noyau RBF :", grid_search_rbf.best_params_)

# Choix du meilleur modle pour le noyau RBF
best_svm_rbf_model = grid_search_rbf.best_estimator_

# Prdiction sur l'ensemble de test avec le meilleur modle pour le noyau RBF
y_pred_rbf = best_svm_rbf_model.predict(X_test)

# valuation de la performance du modle pour le noyau RBF
accuracy_rbf = best_svm_rbf_model.score(X_test, y_test)
print("Prcision pour le noyau RBF :", accuracy_rbf)
```
